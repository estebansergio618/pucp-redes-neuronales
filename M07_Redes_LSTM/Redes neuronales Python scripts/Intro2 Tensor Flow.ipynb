```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = np.linspace(1,10,num = 10)
x
```

```python
y = 2*x**2 + 10

pd.DataFrame({'feature':x
              ,'target':y})
```

```python
plt.plot(x,y,'bo',label ='modelo lineal')
plt.legend()
plt.show()
```

```python
import tensorflow as tf
```

# Escalamiento

```python
from sklearn.preprocessing import MinMaxScaler
scx = MinMaxScaler()
scy = MinMaxScaler()

x=scx.fit_transform(x.reshape(-1,1))
y=scy.fit_transform(y.reshape(-1,1))

print(x,y)
```

```python
h1 = tf.keras.layers.Dense(units=100,input_shape=[1],activation='relu')
h2 = tf.keras.layers.Dense(units=100,activation='relu')
h3 = tf.keras.layers.Dense(units=100,activation='tanh')
h4 = tf.keras.layers.Dense(units=100,activation='sigmoid')

salida = tf.keras.layers.Dense(units = 1)
modelo1 = tf.keras.Sequential([h1,h2,h3,h4,salida])
```

```python
modelo1.compile(
optimizer = tf.keras.optimizers.Adam(0.01),
loss ='mean_squared_error'
)
```

```python
modelo1.summary()
```

```python
historial1 = modelo1.fit(x,y,epochs = 500)
```

```python
plt.xlabel('Epoca')
plt.ylabel('Magnitud loss')
plt.plot(historial1.history['loss'])
```

```python
# y =
resultado1 = modelo1.predict([5])
resultado1
```

```python
print(h1.get_weights())
print(h2.get_weights())
print(salida.get_weights())
```

```python
# plotear los valores reales vs red neuronal
plt.plot(x,y,'bo',label ='modelo lineal')
plt.plot(x,modelo1.predict(x),color='red',label='RNA1')
plt.legend()
plt.show()
```
